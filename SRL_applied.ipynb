```python
import XLM_R as bert
import importlib
from transformers import AutoModelForTokenClassification, AutoTokenizer, Trainer
from transformers import DataCollatorForTokenClassification
import glob
import os
```

```python
importlib.reload(bert)
```

```python
#reload tokenizer and model

tokenizer_xlmr = AutoTokenizer.from_pretrained("../SRL/XLMR_SRL_endu")
xlmr = AutoModelForTokenClassification.from_pretrained("../SRL/XLMR_SRL_endu")
xlmr.resize_token_embeddings(len(tokenizer_xlmr))
data_collator = DataCollatorForTokenClassification(tokenizer_xlmr)
```

```python
ft_trainer = Trainer(
    xlmr,
    processing_class = tokenizer_xlmr,
    data_collator=data_collator,
)
```

```python
for filename in glob.glob('articles_conll/*'):
    base, ext = os.path.splitext(filename)
    l = base.split('/')
    name = l[-1]
    
    duplicated_file = "data/duplicated/" + name + ".dup"
    output_name = "data/srl_pred/" + name + ".pred"

    bert.duplicate_predicates(filename, duplicated_file)
    bert.predict_srl(duplicated_file, ft_trainer, output_name)

```
